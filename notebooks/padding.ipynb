### **Introduction**

This notebook shows how to correct for the region of interest problem present in 3D cone-beam CT scans. It uses one of the scans of the T-rex Casper.

### **Imports**

```python
import qim3d.viz
```

```python
from src import cfg
from scripts.reconstruct import reconstruct_scans
from src.io import load_recon
```

### **Reconstruting the scan**

```python
scan = cfg.SCANS_BY_LAYER[4][2]
print(scan)
```

This is using the raw dataset without padding (c.f. the dataset config descriptions in the `README.md`):

```python
reconstruct_scans(scans=[scan], dataset=cfg.raw)
```

This is using the padding method:

```python
reconstruct_scans(scans=[scan], dataset=cfg.full_res)
```

### **Comparing the reconstructions visually**

```python
vol_raw = load_recon(scan, dataset=cfg.raw)
vol_padded = load_recon(scan, dataset=cfg.full_res)
```

Here we compare the two volumes in an interactive slicer that allows sliding between the volumes to easily see the "before" and "after".

```python
qim3d.viz.overlay(vol_raw, vol_padded)
```

The effect is now more clear. We remove the circular arc artifacts towards the area where the object is excluded, and the intensity gradients that surround the arcs are also removed.

See this notebook for more details about the padding technique: https://github.com/qim-center/qim-notebooks/blob/main/reconstruction/padding.ipynb

But essentially, it simply pads the sinogram in a way that pushes the artifacts outside the ROI.
